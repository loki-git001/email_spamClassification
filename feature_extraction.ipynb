```python
import os
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords

# Downloading NLTK data
nltk.download("stopwords")
nltk.download("punkt")
nltk.download("punkt_tab")

pd.set_option("display.max_columns", None)
```

```python
RAW_DATASET_PATH = "dataset/raw/emails.csv"

df = pd.read_csv(RAW_DATASET_PATH, encoding='latin1')

df.head()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates(keep = 'first')
```

```python
df.shape
```

```python
def add_new_features(df):
    df = df.copy()

    df["num_chars"] = df["text"].str.len()
    df["num_words"] = df["text"].apply(lambda x: len(nltk.word_tokenize(x)))
    df["num_sentences"] = df["text"].apply(lambda x: len(nltk.sent_tokenize(x)))

    return df


df = add_new_features(df)
```

```python
df.head()
```

```python
PROCESSED_DATA_PATH = os.path.join(".", "dataset", "processed_data")
os.makedirs(PROCESSED_DATA_PATH, exist_ok=True)
CSV_PATH = os.path.join(PROCESSED_DATA_PATH, "emails_processed.csv")

df.to_csv(CSV_PATH, index=False)
```
